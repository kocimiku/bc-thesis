## Final models for each dataset
- In this section we will compare performance of two models on unseeen data for each dataset.
- For each comparison we selected the interpretable model with the best performance on the validation set and compare it with the black-box model.

### South German credit dataset
- Since all the interpretable models achieved the same performance in terms of F1 score, we chose to use the SLIM model, since it is the only dataset where it achieves the best performance out of the interpretable models.

```python
%%capture
Xtrain, Xval, Xtest, ytrain, yval, ytest = load_credit()
X_final = pd.concat([Xtrain, Xval])
y_final = pd.concat([ytrain, yval])
c2, conf2, rep2 = train_final_cat(cat_c["params"], X_final, Xtest, y_final, ytest)
c1, conf1, rep1 = train_final_slim(slim_c["params"], slim_c["weights"], X_final, Xtest, y_final, ytest)
```

```python
print(rep1)
```

```python
print(rep2)
```

- In this comparison, the SLIM performs significatly better in terms of F1 score (0.57 vs 0.45) and recall (0.61 vs 0.35), but CatBoost provides slightly better precion (0.54 vs 0.62).

```python
show_confusion_matrix(conf1, conf2, ["high risk", "low risk"], "../figs/credit.pdf")
```

- The confusion matrix on the left belongs to the SLIM, and the one on the right to the CatBoost classifier.
- SLIM is more likely to predict that a client is high risk even though he is not, nearly half of the high risk predictions are wrong.
- CatBoost produces a lot of false negatives, with significatly less false positives, when compared to the SLIM.
- In the context of the domain, we would probably rather predict that a client is high risk than low risk, so the SLIM model would be the obvious choice here. 

### Heart disease dataset
- The interpretable model we will use for this dataset will be the EBM.

```python
Xtrain, Xval, Xtest, ytrain, yval, ytest = load_heart()
X_final = pd.concat([Xtrain, Xval])
y_final = pd.concat([ytrain, yval])
hr1, conf1, rep1 = train_final_gam(gam_hr["params"], X_final, Xtest, y_final, ytest)
hr2, conf2, rep2 = train_final_cat(cat_hr["params"], X_final, Xtest, y_final, ytest)
```

```python
print(rep1)
```

```python
print(rep2)
```

- In this case, the CatBoost classifier performs very slightly better in each performance metric compared to the EBM (precision - 0.80 vs 0.81, recall - 0.88 vs 0.90, f1 - 0.83 vs 0.85).   

```python
show_confusion_matrix(conf1, conf2, ["sick", "healthy"], "../figs/heart.pdf")
```

- The confusion matrix on the left belongs to the EBM, and the one on the right to the CatBoost.
- Both of the matrices are nearly the same, with both classifiers having a near identical ratio of false positives and false negatives.
- The improvement in performance when using a black-box is minimal in this case, which we believe is far outweighted by the interpretability of the EBM clasifier.

### Hypothyroid dataset
- The interpretable model we will use for this dataset is the GOSDT classifier.

```python
%%capture
Xtrain, Xval, Xtest, ytrain, yval, ytest = load_hypo_nonan()
X_final = pd.concat([Xtrain, Xval])
y_final = pd.concat([ytrain, yval])
hp1, conf1, rep1 = train_final_gosdt(odt_hp["params"], X_final, Xtest, y_final, ytest)
Xtrain, Xval, Xtest, ytrain, yval, ytest = load_hypothyroid()
X_final = pd.concat([Xtrain, Xval])
y_final = pd.concat([ytrain, yval])
hp2, conf2, rep2 = train_final_cat(cat_hp["params"], X_final, Xtest, y_final, ytest)
```

- The confusion matrix on the left belongs to the GOSDT, and the one on the right to the CatBoost.
- As we can see there is only one case, where Catboost correctly classifies the data point as positive, while the GOSDT predicts negative.
- Such an improvement is not very significant, so once again we feel as tough it is better to use the GOSDT for this task.
